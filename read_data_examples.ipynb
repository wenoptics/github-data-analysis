```python
import pandas as pd
import json
```

```python
def read_custom_json(file):
    s = filter(lambda i: i.strip()!='', file.read().split('\n'))
    return [json.loads(l) for l in s]
```

```python
# repo_basics
```

```python
repo_basics = read_custom_json(open('data/repo_basics.json', 'r'))
```

```python
df_repo_basics = pd.DataFrame(repo_basics)
df_repo_basics.set_index('full_name', inplace=True)
df_repo_basics.head()
```

```python
# repo_commits
```

```python
repo_commits = read_custom_json(open('data/repo_commits.json', 'r'))
df_repo_commits = pd.DataFrame(repo_commits)
df_repo_commits.set_index('repo_fullname', inplace=True)
df_repo_commits.head()
```

```python
## repo_pulls
```

```python
repo_pulls = read_custom_json(open('data/repo_pulls.json', 'r'))
df_repo_pulls = pd.DataFrame(repo_pulls)
df_repo_pulls.set_index('repo_fullname', inplace=True)
df_repo_pulls.head()
```
